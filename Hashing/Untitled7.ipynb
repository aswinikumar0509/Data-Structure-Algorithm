```python
def intersection(a,b,n,m):
    i,j,k = 0,0,0
    result = [0 for i in range(n+m)]
    while i<n and j<m:
        if a[i]<b[j]:
            i+=1
        elif a[i]>b[j]:
            j+=1
        else:
            if k!=0 and a[i]==result[k-1]:
                i+=1
                j+=1
            else:
                result[k] = a[i]
                i+=1
                j+=1
                k+=1
                
    print("Intersection",*result[:k])     
```
