```python
### Binary Search Algorithm

def binarySearch(arr,l,r,x):
    while l<=r:
        mid = l+(r-l)//2
        
        if arr[mid]==x:
            return mid
        elif arr[mid]<x:
            l = mid+1
        else:
            r = mid-1
            
    return -1

arr = [2,3,4,10,40]
x = 10
l = 0
r = len(arr)-1
binarySearch(arr,l,r,10)
```

```python
## FInding the first occurance of the element 

def firstOccur(arr,x):
    for i in range(len(arr)):
        if arr[i]==x:
            return i
    return -1

arr = [10,20,30,40,50,50,50]
x = 50
firstOccur(arr,x)

#time = O(N)
```

```python
def first(arr,low,high,x,n):
    if(high>=low):
        mid = low+(high-low)//2
        if((mid==0 or x > arr[mid-1]) and arr[mid]==x):
            return mid
        elif (x > arr[mid]):
            return first(arr,(mid+1),high,x,n)
        else:
            return first(arr,low,(mid-1),x,n)
        
    return -1

def last(arr,low,high,x,n):
    if(high>=low):
        mid = low+(high-low)//2
        if((mid==n-1 or x< arr[mid+1]) and arr[mid]==x):
            return mid
        elif (x < arr[mid]):
            return last(arr,low,(mid-1),x,n)
        else:
            return last(arr,(mid+1),high,x,n)
        
    return -1

arr = [1,2,2,2,2,3,4,7,8,8]
n = len(arr)

x = 8
print("First Occurence = " , first(arr,0,n-1,x,n))
print("Last Occurance = ", last(arr,0,n-1,x,n))
            
```
