```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key


def finmax(root):
    if(root==None):
        return float("-inf")
    
    res = root.val

    lh = finmax(root.left)
    rh = finmax(root.right)

    if(lh>rh):
        res = lh
    if(rh>lh):
        res = rh

    return res

def Min(root):

    if root is None:
        return float('inf')  

    return min(root.val, Min(root.left), Min(root.right))

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

print(finmax(root))
print(Min(root))
```

```python
class Node:

    def __init__(self,key):
        self.right = None
        self.left = None
        self.key = key


def getSucc(curr):
    while curr.left!=None:
        curr = curr.left

    return curr.key

def deletNode(root,key):

    if root == None:
        return None
    
    if root.key>key:
        root.left = deletNode(root.left,key)
    elif root.key<key:
        root.right = deletNode(root.right,key)

    else:
        if root.left==None:
            return root.right
        elif root.right==None:
            return root.left
        else:
            succ = getSucc(root.right)
            root.key = succ
            root.right = deletNode(root.right,succ)

    return root


def inorder(root):
    if root:
        inorder(root.left)
        print(root.key,end= " ")
        inorder(root.right)

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)


node = deletNode(root,40)
inorder(node)
```

```python
### Operation on the heap

from heapq import heappush,heappop,heapify

class Minheap:

    def __init__(self):
        self.heap = []

    def parent(self,i):
        return (i-1)/2
    
    def insertKey(self,k):
        heappush(self.heap,k)

    def decreasekey(self,i,new_val):
        self.heap[i] = new_val
        while(i!=0 and self.heap[self.parent(i)]>self.heap[i]):
            self.heap[i],self.heap[self.parent(i)] = (self.heap[self.parent(i)],self.heap(i))

    def extractMin(self):
        return heappop(self.heap)
    
    def deletekey(self,i):
        self.decreasekey(i,float("-inf"))
        self.extractMin()

    def getMin(self):
        return self.heap[0]
    

heapObj = Minheap()
heapObj.insertKey(3)
heapObj.insertKey(2)
# heapObj.deletekey(1)
heapObj.insertKey(15)
heapObj.insertKey(5)
heapObj.insertKey(4)
heapObj.insertKey(45)
  
print(heapObj.extractMin()),
print(heapObj.getMin())
# heapObj.decreaseKey(2, 1)
print(heapObj.getMin())

```

```python
### Find the subsequence of the array

def isSubseq(s1,s2):
    i,j=0,0
    if len(s2)>len(s1):
        return False
    while (i<len(s1) and j < len(s2)):
        if s1[i]==s2[j]:
            j=j+1
        i+=1

    if j==len(s2):
        return True
    else:
        return False
        
s1 = "ABCDEFG"
s2 = "ADE"
isSubseq(s1,s2)
```

```python
### Finding the anagram of the string

def areAnagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    count = [0]*256
    for i in range(len(s1)):
        count[ord(s1[i])]+=1
        count[ord(s2[i])]-=1
    for x in count:
        if x!=0:
            return False
    return True

s1 = "aabca"
s2 = "acaba"
areAnagram(s1,s2)
```

```python
### Reversing of the word

def reverse(s, b, e):
    while b < e:
        s[b], s[e] = s[e], s[b]
        b += 1
        e -= 1

def reverseWord(s):
    s = list(s)
    n = len(s)
    
    b = 0
    for i in range(n):
        if s[i] == ' ':
            reverse(s, b, i - 1)
            b = i + 1
    
    reverse(s, b, n - 1)
    reverse(s, 0, n - 1)
    
    return ''.join(s)
s = " I love Coding"
reverseWord(s)
```

```python
## Stock Spam problem

def printSpam(arr):

    for i in range(len(arr)):
        spam=1
        j=i-1
        while j>=0 and arr[i]>arr[j]:
            spam+=1
            j-=1
        print(spam,end=" ")

arr = [100,80,90,120]
printSpam(arr)
```

```python
### Determining the tree is balanced or not
class Node:

    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None


def isBalanaced(root):

    if root is None:
        return True
    
    lh = isBalanaced(root.left)
    if lh==-1:
        return -1
    
    rh = isBalanaced(root.right)
    if rh==-1:
        return -1
    
    if (abs(lh-rh)>1):
        return -1
    else:
        return max(lh,rh)+1
    



root = Node(10)
root.left = Node(5)
root.right = Node(30)
root.right.left = Node(15)
root.right.right = Node(20)
if(isBalanaced(root) == -1):
    print("Not Balabced")
else:
    print("Balanaced")
```
