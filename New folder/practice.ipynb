```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key


def printInorder(root):

    if root:
        printInorder(root.left)
        print(root.val)
        printInorder(root.right)


def printpreOrder(root):
    if root:
        print(root.val)
        printpreOrder(root.left)
        printpreOrder(root.right)

def printpostOrder(root):
    if root:
        printpostOrder(root.left)
        printpostOrder(root.right)
        print(root.val)




if __name__=="__main__":

    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)


    printInorder(root)
    print("=================")
    printpostOrder(root)
    print("=================")
    printpreOrder(root)
```

```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key

def height(root):

    if root==None:
        return 0
    
    lh = height(root.left)
    rh = height(root.right)

    return max(lh,rh)+1


####### K distance from the root

def printKdistance(root,k):
    if root==None:
        return
    
    if k==0:
        print(root.val)
    else:
        printKdistance(root.left,k-1)
        printKdistance(root.right,k-1)

if __name__=="__main__":

    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.right = Node(7)
    root.right.left = Node(6)

    print(height(root))
    printKdistance(root,2)
```

```python
### level order traversal of the tree

class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key

def levelorderTraversal(root):

    if root is None:
        return
    
    queue = []
    queue.append(root)

    while(len(queue)>0):
        print(queue[0].val,end=" ")
        node = queue.pop(0)

        if node.left is not None:
            queue.append(node.left)
        if node.right is not None:
            queue.append(node.right)

def size(root):
    if root is None:
        return 0
    else:
        return (size(root.left) + 1 + size(root.right))


root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

levelorderTraversal(root)
print(size(root))
```

```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key


def finmax(root):
    if(root==None):
        return float("-inf")
    
    res = root.val

    lh = finmax(root.left)
    rh = finmax(root.right)

    if(lh>rh):
        res = lh
    if(rh>lh):
        res = rh

    return res

def Min(root):

    if root is None:
        return float('inf')  

    return min(root.val, Min(root.left), Min(root.right))

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

print(finmax(root))
print(Min(root))
```

```python
class Node:

    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data

def search(root,data):

    if root is None:
        return None
    
    if root.data==data:
        return root
    
    left = search(root.left,data)
    if left:
        return left
    
    return search(root.left,data)
    
root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

node=search(root,20)
print(node.data if node else None)
```

```python
class Node:

    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data


def insert(root,data):
    if root is None:
        return Node(data)
    
    else:
        if root.data==data:
            return root
        elif root.data<data:
            root.right = insert(root.right,data)
        else:
            root.left = insert(root.left,data)

    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end= " ")
        inorder(root.right)

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

node=insert(root,100)
inorder(node)
```

```python
class Node:

    def __init__(self,key):
        self.right = None
        self.left = None
        self.key = key


def getSucc(curr):
    while curr.left!=None:
        curr = curr.left

    return curr.key

def deletNode(root,key):

    if root == None:
        return None
    
    if root.key>key:
        root.left = deletNode(root.left,key)
    elif root.key<key:
        root.right = deletNode(root.right,key)

    else:
        if root.left==None:
            return root.right
        elif root.right==None:
            return root.left
        else:
            succ = getSucc(root.right)
            root.key = succ
            root.right = deletNode(root.right,succ)

    return root


def inorder(root):
    if root:
        inorder(root.left)
        print(root.key,end= " ")
        inorder(root.right)

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)


node = deletNode(root,40)
inorder(node)
```

```python
### Flore value of the BST


INT_MAX = 21498579375

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insert(root, data):
    if root is None:
        return Node(data)

    if data < root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)

    return root

def floor(root,data):
    if root == None:
        return INT_MAX
    
    if root.data==data:
        return root.data
    
    if root.data>data:
        return floor(root.left,data)
    
    floorvalue = floor(root.right,data)

    return floorvalue if floorvalue<data else root.data


if __name__ == '__main__':

    root = None
    root = insert(root, 7) 
    insert(root, 10) 
    insert(root, 5) 
    insert(root, 3) 
    insert(root, 6) 
    insert(root, 8) 
    insert(root, 12) 
    print(floor(root, 9))

```
