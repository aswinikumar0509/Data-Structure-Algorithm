```python
### Find the subsequence of the array

def isSubseq(s1,s2):
    i,j=0,0
    if len(s2)>len(s1):
        return False
    while (i<len(s1) and j < len(s2)):
        if s1[i]==s2[j]:
            j=j+1
        i+=1

    if j==len(s2):
        return True
    else:
        return False
        
s1 = "ABCDEFG"
s2 = "ADE"
isSubseq(s1,s2)
```

```python
### Finding the anagram of the string

def areAnagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    count = [0]*256
    for i in range(len(s1)):
        count[ord(s1[i])]+=1
        count[ord(s2[i])]-=1
    for x in count:
        if x!=0:
            return False
    return True

s1 = "aabca"
s2 = "acaba"
areAnagram(s1,s2)
```

```python
### Reversing of the word

def reverse(s, b, e):
    while b < e:
        s[b], s[e] = s[e], s[b]
        b += 1
        e -= 1

def reverseWord(s):
    s = list(s)
    n = len(s)
    
    b = 0
    for i in range(n):
        if s[i] == ' ':
            reverse(s, b, i - 1)
            b = i + 1
    
    reverse(s, b, n - 1)
    reverse(s, 0, n - 1)
    
    return ''.join(s)
s = " I love Coding"
reverseWord(s)
```

```python
### Determining the tree is balanced or not
class Node:

    def __init__(self,data):
        self.data = data
        self.left = None
        self.right = None


def isBalanaced(root):

    if root is None:
        return True
    
    lh = isBalanaced(root.left)
    if lh==-1:
        return -1
    
    rh = isBalanaced(root.right)
    if rh==-1:
        return -1
    
    if (abs(lh-rh)>1):
        return -1
    else:
        return max(lh,rh)+1
    



root = Node(10)
root.left = Node(5)
root.right = Node(30)
root.right.left = Node(15)
root.right.right = Node(20)
if(isBalanaced(root) == -1):
    print("Not Balabced")
else:
    print("Balanaced")
```

```python
# Fractional Knapsack

def fknaps(w,arr) :
    
    n = len(arr)
    costweight = []
    
    for i in range(n) :
        c,W = arr[i][0], arr[i][1]
        costweight.append((c, W, ((c*1.0)/w)))
        
    costweight = sorted(costweight, key=lambda x : x[2],reverse = True)
    
    res = 0.0
    
    for curr in costweight :
        
        if curr[1] <= w :
            res += curr[0]
            w -= curr[1]
            
        else :
            res += (curr[0] * (w/curr[1]))
            break
    
    return res 
    
    
arr = [(120,30),(100,20),(60,10)]
w = 50
print(fknaps(w,arr))
```

```python
### tabulation method

def lcs(s1,s2):
    m=len(s1)
    n=len(s2)
    dp = [[None]*(n+1) for i in range(m+1)]
    for i in range(m+1):
        for j in range(n+1):
            if i==0 or j==0:
                dp[i][j]=0
            elif s1[i-1]==s2[j-1]:
                dp[i][j] = 1+ dp[i-1][j-1]
            else:
                dp[i][j] = max(dp[i-1][j],dp[i][j-1])

    return dp[m][n]
s1 = "abeds"
s2= "wbhgs"
n=len(s1)
m=len(s2)
print(lcs(s1,s2))
```

```python
### number of way to find the sum of the coin

def countWays(coin,n,s):
    dp = [[0 for i in range(s+1)] for i in range(n+1)]
    for i in range(n+1):
        dp[i][0] = 1

    for i in range(1,n+1):
        for j in range(1,s+1):
            dp[i][j] = dp[i-1][j]
            if j>=coin[i-1]:
                dp[i][j]+=dp[i][j-coin[i-1]]

    return dp[n][s]

coin = [1,2,3]
n = len(coin)
s = 4
countWays(coin,n,s)
```

```python
### Way to reach the stair

def ways(n):
    dp = [0]*(n+1)
    dp[0]=1
    dp[1]=1
    for i in range(2,n+1):
        dp[i]=dp[i-1]+dp[i-2]

    return dp[n]

n=4
ways(n)
```

```python
d, sumTime = map(int, input().split())

mins = []
maxs = []
for _ in range(d):
    mn, mx = map(int, input().split())
    mins.append(mn)
    maxs.append(mx)

sumMin = sum(mins)
sumMax = sum(maxs)


if sumTime < sumMin or sumTime > sumMax:
    print("NO")
else:
    print("YES")
    schedule = mins[:]  
    remaining = sumTime - sumMin

    for i in range(d):
        
        can_add = maxs[i] - mins[i]
        add = min(remaining, can_add)
        schedule[i] += add
        remaining -= add
        if remaining == 0:
            break

    print(*schedule)
```
