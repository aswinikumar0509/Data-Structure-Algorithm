```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key


def printInorder(root):

    if root:
        printInorder(root.left)
        print(root.val)
        printInorder(root.right)


def printpreOrder(root):
    if root:
        print(root.val)
        printpreOrder(root.left)
        printpreOrder(root.right)

def printpostOrder(root):
    if root:
        printpostOrder(root.left)
        printpostOrder(root.right)
        print(root.val)




if __name__=="__main__":

    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)


    printInorder(root)
    print("=================")
    printpostOrder(root)
    print("=================")
    printpreOrder(root)
```

```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key

def height(root):

    if root==None:
        return 0
    
    lh = height(root.left)
    rh = height(root.right)

    return max(lh,rh)+1


####### K distance from the root

def printKdistance(root,k):
    if root==None:
        return
    
    if k==0:
        print(root.val)
    else:
        printKdistance(root.left,k-1)
        printKdistance(root.right,k-1)

if __name__=="__main__":

    root = Node(1)
    root.left = Node(2)
    root.right = Node(3)
    root.left.left = Node(4)
    root.left.right = Node(5)
    root.right.right = Node(7)
    root.right.left = Node(6)

    print(height(root))
    printKdistance(root,2)
```

```python
### level order traversal of the tree

class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key

def levelorderTraversal(root):

    if root is None:
        return
    
    queue = []
    queue.append(root)

    while(len(queue)>0):
        print(queue[0].val,end=" ")
        node = queue.pop(0)

        if node.left is not None:
            queue.append(node.left)
        if node.right is not None:
            queue.append(node.right)

def size(root):
    if root is None:
        return 0
    else:
        return (size(root.left) + 1 + size(root.right))


root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

levelorderTraversal(root)
print(size(root))
```

```python
class Node:

    def __init__(self,key):
        self.left = None
        self.right = None
        self.val = key


def finmax(root):
    if(root==None):
        return float("-inf")
    
    res = root.val

    lh = finmax(root.left)
    rh = finmax(root.right)

    if(lh>rh):
        res = lh
    if(rh>lh):
        res = rh

    return res

def Min(root):

    if root is None:
        return float('inf')  

    return min(root.val, Min(root.left), Min(root.right))

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

print(finmax(root))
print(Min(root))
```

```python
class Node:

    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data

def search(root,data):

    if root is None:
        return None
    
    if root.data==data:
        return root
    
    left = search(root.left,data)
    if left:
        return left
    
    return search(root.left,data)
    
root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

node=search(root,20)
print(node.data if node else None)
```

```python
class Node:

    def __init__(self,data):
        self.left = None
        self.right = None
        self.data = data


def insert(root,data):
    if root is None:
        return Node(data)
    
    else:
        if root.data==data:
            return root
        elif root.data<data:
            root.right = insert(root.right,data)
        else:
            root.left = insert(root.left,data)

    return root

def inorder(root):
    if root:
        inorder(root.left)
        print(root.data,end= " ")
        inorder(root.right)

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)

node=insert(root,100)
inorder(node)
```

```python
class Node:

    def __init__(self,key):
        self.right = None
        self.left = None
        self.key = key


def getSucc(curr):
    while curr.left!=None:
        curr = curr.left

    return curr.key

def deletNode(root,key):

    if root == None:
        return None
    
    if root.key>key:
        root.left = deletNode(root.left,key)
    elif root.key<key:
        root.right = deletNode(root.right,key)

    else:
        if root.left==None:
            return root.right
        elif root.right==None:
            return root.left
        else:
            succ = getSucc(root.right)
            root.key = succ
            root.right = deletNode(root.right,succ)

    return root


def inorder(root):
    if root:
        inorder(root.left)
        print(root.key,end= " ")
        inorder(root.right)

root = Node(10)
root.left = Node(20)
root.right = Node(30)
root.left.left = Node(40)
root.right.left = Node(50)
root.right.right = Node(60)
root.right.left.left = Node(70)
root.right.left.right = Node(80)


node = deletNode(root,40)
inorder(node)
```

```python
### Flore value of the BST


INT_MAX = 21498579375

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def insert(root, data):
    if root is None:
        return Node(data)

    if data < root.data:
        root.left = insert(root.left, data)
    else:
        root.right = insert(root.right, data)

    return root

def floor(root,data):
    if root == None:
        return INT_MAX
    
    if root.data==data:
        return root.data
    
    if root.data>data:
        return floor(root.left,data)
    
    floorvalue = floor(root.right,data)

    return floorvalue if floorvalue<data else root.data


if __name__ == '__main__':

    root = None
    root = insert(root, 7) 
    insert(root, 10) 
    insert(root, 5) 
    insert(root, 3) 
    insert(root, 6) 
    insert(root, 8) 
    insert(root, 12) 
    print(floor(root, 9))

```

```python
### Operation on the heap

from heapq import heappush,heappop,heapify

class Minheap:

    def __init__(self):
        self.heap = []

    def parent(self,i):
        return (i-1)/2
    
    def insertKey(self,k):
        heappush(self.heap,k)

    def decreasekey(self,i,new_val):
        self.heap[i] = new_val
        while(i!=0 and self.heap[self.parent(i)]>self.heap[i]):
            self.heap[i],self.heap[self.parent(i)] = (self.heap[self.parent(i)],self.heap(i))

    def extractMin(self):
        return heappop(self.heap)
    
    def deletekey(self,i):
        self.decreasekey(i,float("-inf"))
        self.extractMin()

    def getMin(self):
        return self.heap[0]
    

heapObj = Minheap()
heapObj.insertKey(3)
heapObj.insertKey(2)
# heapObj.deletekey(1)
heapObj.insertKey(15)
heapObj.insertKey(5)
heapObj.insertKey(4)
heapObj.insertKey(45)
  
print(heapObj.extractMin()),
print(heapObj.getMin())
# heapObj.decreaseKey(2, 1)
print(heapObj.getMin())

```

```python
## Checking the string is rotated or not


def are_rotated(s1,s2):
    if len(s1)!=len(s2):
        return False
    temp = s1+s2
    return temp.find(s2)!=-1

s1 = "ABDC"
s2 = "EFGHI"
print(are_rotated(s1,s2))
```

```python
### Find the subsequence of the array

def isSubseq(s1,s2):
    i,j=0,0
    if len(s2)>len(s1):
        return False
    while (i<len(s1) and j < len(s2)):
        if s1[i]==s2[j]:
            j=j+1
        i+=1

    if j==len(s2):
        return True
    else:
        return False
        
s1 = "ABCDEFG"
s2 = "ADE"
isSubseq(s1,s2)
```

```python
### Finding the anagram of the string

def areAnagram(s1,s2):
    if len(s1)!=len(s2):
        return False
    count = [0]*256
    for i in range(len(s1)):
        count[ord(s1[i])]+=1
        count[ord(s2[i])]-=1
    for x in count:
        if x!=0:
            return False
    return True

s1 = "aabca"
s2 = "acaba"
areAnagram(s1,s2)
```
